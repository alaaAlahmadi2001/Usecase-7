## 4. Bivariate/Multivariate Analysis:


#### 1. Which positions have the highest average current market value?

```python
# Attacking positions like Second Striker and Left Winger have the highest average market values, followed closely by central and attacking midfielders.
#In contrast, defensive roles and goalkeepers tend to be valued lower,
#highlighting the premium placed on offensive and playmaking positions in the market.
```

```python
average_market_value = df.groupby('position')['current_value'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
average_market_value.plot(kind='bar', color='lightblue')
plt.title('Average Current Market Value by Position')
plt.xlabel('Position')
plt.ylabel('Average Current Market Value (EUR)')
plt.show()
```

### 2. How does the number of days injured impact the number of appearances?

```python
#The scatter plot shows that as the number of days injured increases, the number of appearances tends to decrease.
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['days_injured'], df['appearance'], c=df['position_encoded'], cmap='coolwarm', alpha=0.6)
plt.colorbar(label='Position Encoded')
plt.title('Impact of Days Injured on Number of Appearances')
plt.xlabel('Days Injured')
plt.ylabel('Number of Appearances')
plt.show()
```

### 3. Is there a correlation between the number of goals scored and the player's age?

```python
#There’s no clear link between age and the number of goals scored. Most players score between 0 and 2 goals,
#regardless of age, suggesting that other factors influence goal-scoring more than age.
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['goals'], c=df['position_encoded'], cmap='viridis', alpha=0.6)
plt.colorbar(label='Position Encoded')
plt.title('Correlation between Age and Number of Goals')
plt.xlabel('Age')
plt.ylabel('Number of Goals')
plt.show()
```

#### 2. Feature engineering

#### 3. Feature selection

```python
correlation = df.corr()
print(correlation['current_value'].sort_values(ascending=False))
```

```python
# Set the correlation threshold
threshold = 0.25  
selected_features = correlation[abs(correlation['current_value']) > threshold]['current_value'].index
selected_features
```

```python
selected_features = ['appearance', 'minutes played', 'award','current_value','highest_value']
```

```python
df = df[selected_features]
```

```python
df.head()
```

#### 4. Prepare train and test data

```python
# Prepare data
X = df.drop(['current_value'], axis=1)
y = df['current_value']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    random_state=42)

# sacle the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X.shape
```

## Buliding the Model

```python
model = LinearRegression()
```

## Train the Model

```python
model.fit(X_train_scaled, y_train)
```

## Test the Model

```python
y_pred = model.predict(X_test_scaled)
```

## Interpretation of the Model

```python
model.coef_
```

```python
coeff_df = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
print(model.intercept_) 
```

```python
from sklearn.linear_model import Lasso

# Define the model
lasso_model = Lasso(alpha=0.1)

# Train the model
lasso_model.fit(X_train, y_train)
```

## Evaluating the Model 

### 1. Mean Square Error

```python
y_base = [y_train.mean()] * len(y_test)
```

```python
mse_base = mean_squared_error(y_test, y_base )
print(f'Mean Squared Error of base model: {mse_base}')
```

```python
mse_train = mean_squared_error(y_train, model.predict(X_train_scaled))
mse_test = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error for training: {mse_train}')
print(f'Mean Squared Error for testing: {mse_test}')
```

### 2. Mean Absolute Error:

```python
mse_base = mean_absolute_error(y_test, y_base )
print(f'Mean Absolute  Error of base model: {mse_base}')
```

```python
mae_train = mean_absolute_error(y_train, model.predict(X_train_scaled))
mae_test = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error for training: {mae_train}')
print(f'Mean Absolute Error for testing: {mae_test}')
```

### 3. Root Mean Squared Error

```python
print(f'Root Mean Squared  Error of base model: {np.sqrt(mse_base)}')
```

```python
print(f'Root Mean Squared Error for training: {np.sqrt(mse_train)}')
print(f'Root Mean Squared Error for testing: {np.sqrt(mse_test)}')
```

### 4. R Square

```python
r2_score_train = r2_score(y_train, model.predict(X_train_scaled))
r2_score_test = r2_score(y_test, y_pred)
print(f'R Square for training: {r2_score_train}')
print(f'R Square for testing: {r2_score_test}')
```
